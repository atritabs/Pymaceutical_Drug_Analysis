```python
# Dependencies and Setup

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
from scipy.stats import linregress

# Study data files

mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results

mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame

combined_data_df = study_results.join(mouse_metadata.set_index('Mouse ID'), on='Mouse ID')

# Display the data table for preview

combined_data_df.head()
```

```python
# Checking the number of mice.

Total_mice_count = combined_data_df["Mouse ID"].value_counts()
print(len(Total_mice_count))
```

## Ensure that the data is uniquely identified

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

duplicate_mice_ID = combined_data_df.loc[combined_data_df.duplicated(['Mouse ID', 'Timepoint'])]['Mouse ID'].unique()
duplicate_mice_ID
```

```python
# Optional: Get all the data for the duplicate mouse ID. 

duplicate_mice_df = combined_data_df.loc[combined_data_df["Mouse ID"] == "g989", :]
duplicate_mice_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.

mouse_clean_df = combined_data_df[combined_data_df['Mouse ID'].isin(duplicate_mice_ID)==False]
mouse_clean_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.

mouse_cleaned_values = mouse_clean_df["Mouse ID"].value_counts()
print(len(mouse_cleaned_values))
```

## Summary Statistics

```python
# Generating a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 

grouped_drug_df=mouse_clean_df.groupby(['Drug Regimen'])

# mean, median, variance, standard deviation, and SEM of the tumor volume. 

#mean
mean=grouped_drug_df['Tumor Volume (mm3)'].mean()
#median
median=grouped_drug_df['Tumor Volume (mm3)'].median()
#variance
var=grouped_drug_df['Tumor Volume (mm3)'].var()
#std
std=grouped_drug_df['Tumor Volume (mm3)'].std()
#sem
sem=grouped_drug_df['Tumor Volume (mm3)'].sem()

# Assemble the resulting series into a single summary DataFrame.

summary_stats = pd.DataFrame({"Mean Tumor Volume":mean, 
                            "Median Tumor Volume":median, 
                           "Tumor Volume Variance":var, 
                           "Tumor Volume Std. Dev.":std, 
                           "Tumor Volume Std. Err.":sem})

# Displaying the Summary statistics table grouped by the 'Drug Regimen' column

summary_stats
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line

agg_summary =  mouse_clean_df.groupby(['Drug Regimen'])[['Tumor Volume (mm3)']].agg(['mean', 'median', 'var', 'std', 'sem'])
agg_summary
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse

greatest_timepoint=mouse_clean_df.groupby("Mouse ID")
greatest_timepoint=greatest_timepoint["Timepoint"].max()
greatest_timepoint_df=pd.DataFrame(greatest_timepoint)

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint

tumor_volume_df=pd.merge(greatest_timepoint_df, mouse_clean_df, on = ["Mouse ID", "Timepoint"])
tumor_volume_df
```

```python
# Put treatments into a list for a for loop (and later for plot labels).

treatments=["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)

tumor_volume_capomulin = []
tumor_volume_ramicane = []
tumor_volume_infubinol = []
tumor_volume_ceftamin = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 

for x in range(len(tumor_volume_df)-1):
    if tumor_volume_df.loc[x, "Drug Regimen"] == "Capomulin":
        tumor_volume_capomulin.append(tumor_volume_df.loc[x, "Tumor Volume (mm3)"])
    if tumor_volume_df.loc[x, "Drug Regimen"] == "Ramicane":
        tumor_volume_ramicane.append(tumor_volume_df.loc[x, "Tumor Volume (mm3)"])
    if tumor_volume_df.loc[x, "Drug Regimen"] == "Infubinol":
        tumor_volume_infubinol.append(tumor_volume_df.loc[x, "Tumor Volume (mm3)"])
    if tumor_volume_df.loc[x, "Drug Regimen"] == "Ceftamin":
        tumor_volume_ceftamin.append(tumor_volume_df.loc[x, "Tumor Volume (mm3)"])

# List of four tumor volumes for the main list to be used in the for loop
        
tumor_volumes=[tumor_volume_capomulin, tumor_volume_ramicane, tumor_volume_infubinol, tumor_volume_ceftamin]

outlier = []
for x in range(len(tumor_volumes)):

# Locate the rows which contain mice on each drug and get the tumor volumes

    quartiles = pd.Series(tumor_volumes[x]).quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    print(f"\n{treatments[x]}:")  
    
    # Locate the rows which contain mice on each drug and get the tumor volumes

    
    # add subset 

    
    # Determine outliers using upper and lower bounds
    
    print(f"The lower quartile of tumor result is: {lowerq}")
    print(f"The upper quartile of tumor result is: {upperq}")
    print(f"The interquartile range of tumor result is: {iqr}")
    print(f"The the median of tumor result is: {quartiles[0.5]}")

    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
                          
    print(f"Values below {lower_bound} could be outliers.")
    print(f"Values above {upper_bound} could be outliers.")
    
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.

fig1, ax1 = plt.subplots(1, figsize=(9, 6))
ax1.set_title("Potential Outlier List for the Top 4 Regimen Treatments")
ax1.set_ylabel("Final Tumor Volume (mm3)")
ax1.boxplot(tumor_volumes)
ax1.set_xticklabels(treatments)
plt.show()
```

## Line and Scatter Plots

```python
# Generate a list of mice treated with Capomulin in order to identify the mouse ID required below

tumor_volume_df.loc[tumor_volume_df["Drug Regimen"] == "Capomulin"].head(10)
```

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin ; choosing Mouse ID: l509 (Top 10 values)

#Create a dataframe with just Capolumin data that is required

capomulin_df_1 = mouse_clean_df.loc[(mouse_clean_df["Drug Regimen"] == "Capomulin") & (mouse_clean_df["Mouse ID"] == "l509") ]
single_capomulin_df = capomulin_df_1[["Timepoint","Tumor Volume (mm3)" ]]
single_capomulin_df = single_capomulin_df.set_index(["Timepoint"])

#get values ready for plotting

timepoint_1 = single_capomulin_df.index.values
tumor_vol_1 = single_capomulin_df.values

#graph build and display

plt.plot(timepoint_1, tumor_vol_1)
plt.title("Capomulin treatment of mouse l509")
plt.xlabel("Timepoint (days)")
plt.ylabel("Tumor Volume (mm3)")
plt.show()

```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen

capomulin_df_2=mouse_clean_df.loc[mouse_clean_df["Drug Regimen"] == "Capomulin"].groupby("Mouse ID")
avg_capomulin_data=capomulin_df_2["Tumor Volume (mm3)"].mean()
mouse_weight_1=capomulin_df_2["Weight (g)"].unique()
plt.scatter(mouse_weight_1, avg_capomulin_data)
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Average Tumor Volume vs. Mouse Weight for Capomulin Regimen Treatment")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

mouse_weight_2 = mouse_weight_1.astype(float)
correlation = st.pearsonr(mouse_weight_2, avg_capomulin_data)
(slope, intercept, rvalue, pvalue, stderr) = linregress(mouse_weight_2, avg_capomulin_data)
regress_values = mouse_weight_2 * slope + intercept

#Indicating the equation of a striaght line as y=mx+b

line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.scatter(mouse_weight_2, avg_capomulin_data)
plt.plot(mouse_weight_2, regress_values, "r-")
plt.annotate(line_eq,(20,36), fontsize=14)
plt.title("Average Tumor Volume vs. Mouse Weight for Capomulin Regimen Treatment")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")

plt.show()

print(f"The correlation coefficient between mouse weight and the average tumor volume is {round(correlation[0], 2)}.")
```
